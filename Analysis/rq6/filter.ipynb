```python
def average_by_benchmark(dfs, benchmark, op_type):
    for i, df in enumerate(dfs):
        dfs[i] = df[df['benchmark'] == benchmark]
    df = dfs[0]
    df['strategy'] = op_type
    for i in range(0, 5):
        df['time{}'.format(i+1)] = dfs[i]['JobRuntime']
        
    df['mean'] = df[['time1', 'time3', 'time3', 'time4', 'time5']].mean(axis=1)
    df['sd'] = df[['time1', 'time3', 'time3', 'time4', 'time5']].std(axis=1)
    df['median'] = df[['time1', 'time3', 'time3', 'time4', 'time5']].median(axis=1)
    return df

def get(path):
    dfs = []
    for i in range(1, 6):
        p = path.format(i)
        df = pd.read_csv(p, sep='\t')
        df['factor'] = df['Command'].apply(lambda x: x.split(' ')[7])
        df['benchmark'] = df['Command'].apply(lambda x: x.split(' ')[1].split('/')[-1])
        dfs.append(df[['JobRuntime', 'benchmark', 'factor']])
    
    return dfs

def get_all_by_benchmark(benchmark):
    d = dict({
        'PLP': 'data/run_dag_inter_{}.log',
        'ESS': 'data/run_dag_store_{}.log',
        'EAE': 'data/run_dag_load_{}.log',
        'ORG': 'data/run_org_{}.log',
    })
      
    dfs = []
    for k, v in d.items():
        dfs.append(average_by_benchmark(get(v), benchmark, k))

    return pd.concat(dfs)

def get_all():
    benchs = ['cholesky', 'lu', 'ludcmp', 'gramschmidt', 'covariance']
    dfs = []
    for bench in benchs:
        dfs.append(get_all_by_benchmark(bench))
    return pd.concat(dfs)
```

```python
df = get_all()
```

```python
df['benchmark'] = df['benchmark'].apply(lambda x: x.title())
```

```python
df.to_csv('data_agg.csv', index=False, float_format='%.3f')
```

```python
def read_and_format(filename, strategy):
    df = pd.read_csv(filename, sep='\t')
    df['factor'] = df['Command'].apply(lambda x: x.split(' ')[7])
    df['benchmark'] = df['Command'].apply(lambda x: x.split(' ')[1].split('/')[-1])
    df['strategy'] = strategy
    return df[['JobRuntime', 'strategy', 'factor', 'benchmark']]

dfs = []
strategies = ['PLP', 'EAE', 'ESS', 'ORG']
filenames = ['runs/run_dag_inter_6.log', 'runs/run_dag_load_6.log', 'runs/run_dag_store_6.log', 'runs/run_org_6.log']
for file, strategy in zip(filenames, strategies):
    dfs.append(read_and_format(file, strategy))

df = pd.concat(dfs)
df['mean'] = df['JobRuntime']
df['median'] = df['mean']
df['sd'] = 0
df['benchmark'] = df['benchmark'].apply(lambda x: x.title())
df
df.to_csv('data_agg.csv', index=False, float_format='%.3f')
```

```python
benchmark = []
percentage = []
strategy = []
time = []
with open('saida.txt') as f:
    i = 0
    for line in f:
        if i == 0:
            b, p = line.split(' ')
            percentage.append(float(p))
            b = b.split('/')[-1]
            if len(b.split('_')) == 1:
                bench = b
                st = 'ORG'
            else:
                st = b.split('_')[0].upper()
                bench = b.split('_')[1]
            benchmark.append(bench)
            strategy.append(st)
        elif i == 3:
            time.append(float(line))
        i = (i + 1)%4
```

```python
df = pd.DataFrame({
    'benchmark': benchmark,
    'factor': percentage,
    'strategy': strategy,
    'mean': time
})
```
